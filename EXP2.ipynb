```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn import metrics
sns.set()
```

```python
iris_data=pd.read_csv('Iris.csv')
iris_data
```

```python
iris_data.info()
```

```python
iris_data.describe()
```

```python
iris_data[iris_data.duplicated()]
```

```python
iris_data['Species'].value_counts()
```

```python
sns.pairplot(iris_data,hue='Species',height=4)
```

```python
numeric_iris_data = iris_data.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr_matrix = numeric_iris_data.corr()

# Plot the heatmap
plt.figure(figsize=(10,11))
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
iris_data.groupby('Species').agg(['mean','median'])
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,9))
sns.boxplot( y='PetalWidthCm', x= 'Species', data=iris_data, orient='v' , ax=axes[0, 0])
sns.boxplot( y='PetalWidthCm', x= 'Species', data=iris_data, orient='v' , ax=axes[0, 1])
sns.boxplot( y='PetalWidthCm', x= 'Species', data=iris_data, orient='v' , ax=axes[1, 0])
sns.boxplot( y='PetalWidthCm', x= 'Species', data=iris_data, orient='v', ax=axes[1, 1])
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,9))
sns.violinplot( y='PetalWidthCm', x= 'Species', data=iris_data, orient='v' , ax=axes[0, 0])
sns.violinplot( y='PetalLengthCm', x= 'Species', data=iris_data, orient='v' , ax=axes[0, 1])
sns.violinplot( y='SepalLengthCm', x= 'Species', data=iris_data, orient='v' , ax=axes[1, 0])
sns.violinplot( y='SepalWidthCm', x= 'Species', data=iris_data, orient='v', ax=axes[1, 1])
plt.show()
```
